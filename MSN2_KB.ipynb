```python
def engines(data):
    value = 0
    if data["ENGINE_RUNNING_1"] is True:
        value += 1 
    if data["ENGINE_RUNNING_2"] is True:
        value += 1 
    if data["ENGINE_RUNNING_3"] is True:
        value += 1 
    if data["ENGINE_RUNNING_4"] is True:
        value += 1 
    return value

file["Num_Engines_Running"] = file.apply(engines,axis=1)
```
